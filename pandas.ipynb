# dropping the unwanted columns
- drop() - drop the columns
- dropna() - drop the rows with missing value
- drop_duplicates() - drop the duplicate rows

```python
# drop
cols_to_drop = ['Type','Coverage','AREA','REG','DEV','dummy_col']
df = df.drop(columns=cols_to_drop)
df
```

```python
df = df.rename(
    columns={
        'OdName':'country',
        'AreaName':'continent',
        'RegName':'region',
        'DevName':'status'
    }
)
df
```

```python
df.sort_values(by='total')
```

```python
df = df.sort_values(by='total',ascending=False) #descending values
df
```

```python
fig , ax = plt.subplots(figsize=(15,6))
sns.barplot(x='country',y='total',data=df.head(10),ax=ax)
plt.title('Top 10 countries with highest immigration overall')
plt.xticks(rotation = 90)
ax.axhline(df.head(10)['total'].mean() ,color ='k',linestyle = '--')
ax.text(5,df.head(10)['total'].mean()+10000,df.head(10)['total'].mean())
plt.show()
```
