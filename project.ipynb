```python
def get_dates():
    url = 'https://www.espn.in/cricket/scores'
    sc2 = Scraper(url)
    t=Tag('header',cls='scoreCollection__PageHeader')
    i=Tag('li',cls='DateCarousel__list__item has-event')
    return sc2.get_repeating_date_from_page(target=t,
                                             items=i,
                                             date=Tag('span',cls='DateCarousel__dayNum'),
                                             day=Tag('span',cls='DateCarousel__dayName'))
```

```python
import pandas as pd
Dates=get_dates()
# Create a DataFrame from the list of dates
pd.DataFrame(Dates)
```
